### Load Environment Variables

```python
import os
import glob
import tiktoken
from dotenv import load_dotenv
from langchain_chroma import Chroma
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.document_loaders import DirectoryLoader, TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import ChatOpenAI
from langchain_core import messages
import gradio as gr
```

### Load API

```python
MODEL = "gpt-4.1-nano"
db_name = "vector_db" #database name for chroma
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
```

### Load all files

```python
knowledge_base_path = "knowledge-base/**/*.md"
files = glob.glob(knowledge_base_path, recursive=True)
print(f"Found {len(files)} files in the knowledge base")
entire_knowledge_base = ""

for file_path in files:
    with open(file_path, 'r', encoding='utf-8') as f:
        entire_knowledge_base += f.read()
        entire_knowledge_base += "\n\n"

print(f"Total characters in knowledge base: {len(entire_knowledge_base):,}")
```

```python
# Check the token
encoding = tiktoken.encoding_for_model(MODEL)
tokens = encoding.encode(entire_knowledge_base)
token_count = len(tokens)
print(f"Total tokens for {MODEL}: {token_count:,}")
```

```python
#load documents
folders = glob.glob("knowledge-base/*")

# view folders in the knowledge base
folders
```

```python
documents = []
for folder in folders:
   # get document name
    doc_type = os.path.basename(folder)
    
    #load files
    loader = DirectoryLoader(
        folder,
        glob="**/*.md", # this is the file type we want to load
        loader_cls=TextLoader, # this is the loader class we want to use
        loader_kwargs={"encoding": "utf-8"} # encoding for the files
    )

    folder_documents = loader.load() # load documents from the folder
    #print(folder_documents) # view loaded documents

    for doc in folder_documents:
        doc.metadata["doc_type"] = doc_type # add document type to metadata
        documents.append(doc) # add document to the list of documents

print(f"Total documents loaded: {len(documents)}")
```

### split to chunks

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=600, chunk_overlap=100) 
chunks = text_splitter.split_documents(documents)
```

```python
print(f"Total chunks created: {len(chunks)}")
```

### Make vectors and store in chroma

```python
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
#create a vector database from the chunks and embeddings
vector_store = Chroma.from_documents(documents=chunks, embedding=embeddings, collection_name=db_name) 
print(f"Total documents in vector database: {vector_store._collection.count()}")
```

```python
# investigate the vectors
collection = vector_store._collection
count = collection.count()

sample_embedding = collection.get(limit=1, include=["embeddings"])["embeddings"][0]
dimensions = len(sample_embedding)
print(f"There are {count:,} vectors with {dimensions:,} dimensions in the vector store")
```

### Search function

```python
def search(question, category=None):

    # If a category is given, use a filter
    if category:
        retriever = vector_store.as_retriever(
            search_kwargs={
                "k": 5,
                "filter": {"doc_type": category}
            }
        )
    else:
        # Search all documents
        retriever = vector_store.as_retriever(
            search_kwargs={"k": 5}
        )

    # Get relevant documents
    results = retriever.invoke(question)
    return results
```

### system prompt

```python
system_prompt = """
You are a helpful support assistant for TechNova.
Answer only using the information below.If the answer is not available,
say:"I don't have information on that in the TechNova knowledge base.

Context:{context}
Question:{question}"""
```

### call LLM

```python
llm = ChatOpenAI(
    model="gpt-4.1-nano",
    temperature=0
)

# Function to get answer
def answer_question(question: str,history):

    # Search relevant documents
    docs = search(question)
    # Create context
    context = ""
    for doc in docs:
        context += doc.page_content + "\n"
        context += "[Source: " + doc.metadata["source"] + "]\n\n"

    messages = [{"role": "system", "content": system_prompt.format(context=context, question=question)}]

    for h in history:
        messages.append({"role": h["role"], "content": h["content"]})
    messages.append({"role": "user", "content": question})

    
    # Get answer from GPT
    response = llm.invoke(messages)
    return response.content
```

### Gradio User Interface

```python
gr.close_all()
gr.ChatInterface(
    answer_question,
    title="TechNova Support Assistant",
    description="Ask questions about TechNova products and policies."
).launch(server_port=7860, prevent_thread_lock=True)
    
```
